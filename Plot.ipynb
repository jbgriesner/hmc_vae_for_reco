```python
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.colors import Normalize
import numpy as np
import matplotlib.cm as cm
from numpy.random import rand

labels = ['recall@5', 'recall@10', 'ndcg@100']

x = np.array([2,2.3, 2.6]) 
width = 0.030  

fs_wrmf= [0.062916, 0.057011, 0.112889]
fs_bpr = [0.012047, 0.021100, 0.021908]
fs_glbavg = [0.000597, 0.000897, 0.000834]
fs_multivae = [0.061, 0.066, 0.127]
fs_metvae = [0.065, 0.064, 0.130]

go_wrmf= [0.062489, 0.089119, 0.103836]
go_bpr = [0.010490, 0.018740, 0.017553] 
go_glbavg = [0.000384, 0.001129, 0.001033]
go_multivae = [0.085, 0.098, 0.153]
go_metvae = [0.087, 0.096, 0.145]

ml_wrmf= [0.125857, 0.194184, 0.197151]
ml_bpr = [0.031608, 0.056786, 0.080661]
ml_glbavg = [0.013343, 0.019863, 0.016347]
ml_multivae = [0.289, 0.303, 0.399]
ml_metvae = [0.307, 0.328, 0.428]

fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20, 5))

my_cmap = cm.get_cmap('jet')
# Get normalize function (takes data in range [vmin, vmax] -> [0, 1])
my_norm = Normalize(vmin=0.0, vmax=0.45) 

ax1.set_title('Results on foursquare')
bar_glbavg_fs = ax1.bar(x - 2*width, fs_glbavg, width, label='GlbAvg', color=my_cmap(my_norm(fs_glbavg)))
bar_bpr_fs = ax1.bar(x - width, fs_bpr, width, label='BPR')
bar_wrmf_fs = ax1.bar(x, fs_wrmf, width, label='WRMF')
bar_multivae_fs = ax1.bar(x + width, fs_multivae, width, label='MultiVAE')
bar_metvae_fs = ax1.bar(x + 2*width, fs_metvae, width, label='MetVAE')
ax1.set_xticks(x)
ax1.set_xticklabels(labels)
ax1.legend()

ax2.set_title('Results on gowalla')
bar_glbavg_go = ax2.bar(x - 2*width, go_glbavg, width, label='GlbAvg', color=my_cmap(my_norm(fs_glbavg)))
bar_bpr_go = ax2.bar(x - width, go_bpr, width, label='BPR')
bar_wrmf_go = ax2.bar(x, go_wrmf, width, label='WRMF')
bar_multivae_go = ax2.bar(x + width, go_multivae, width, label='MultiVAE')
bar_metvae_go = ax2.bar(x + 2*width, go_metvae, width, label='MetVAE')
ax2.set_xticks(x)
ax2.set_xticklabels(labels)
ax2.legend()

ax3.set_title('Results on movielens20M')
bar_glbavg_ml = ax3.bar(x - 2*width, ml_glbavg, width, label='GlbAvg', color=my_cmap(my_norm(fs_glbavg)))
bar_bpr_ml = ax3.bar(x - width, ml_bpr, width, label='BPR')
bar_wrmf_ml = ax3.bar(x, ml_wrmf, width, label='WRMF')
bar_multivae_ml = ax3.bar(x + width, ml_multivae, width, label='MultiVAE')
bar_metvae_ml = ax3.bar(x + 2*width, ml_metvae, width, label='MetVAE')
ax3.set_xticks(x)
ax3.set_xticklabels(labels)
ax3.legend()

# Add some text for labels, title and custom x-axis tick labels, etc.
# ax.set_ylabel('Scores')

fig.tight_layout()

plt.show()
```
